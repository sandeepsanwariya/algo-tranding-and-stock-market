# Algorithmic Trading Strategy Using MACD & Python


```python
from alpha_vantage.timeseries import TimeSeries
import matplotlib.pyplot as plt
import pandas
api_key='YMMGKY70LFXG7QH6'
ts=TimeSeries(key=api_key,output_format='pandas')
data,mata_data=ts.get_intraday('AMD',interval='1min')
data
#print(data['Close'])
```

```python
plt.figure(figsize=(12.5,4.5))
plt.plot(data['4. close'],label='close')
plt.xticks(rotation=45)
plt.xlabel('date')
plt.ylabel("prise")
```

```python
shortEMA=data['4. close'].ewm(span=12,adjust=False).mean()
longEMA=data['4. close'].ewm(span=26,adjust=False).mean()
```

```python
MACD=shortEMA-longEMA
signal=MACD.ewm(span=9,adjust=False).mean()
```

```python
plt.figure(figsize=(12.5,4.5))
plt.plot(data.index,MACD,label='MACD',color='red')
plt.plot(data.index,signal,label='Signal',color='blue')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
```

```python
data['MACD']=MACD
data["Signal_line"]=signal
data
```

```python
def buy_sell(df):
    buy=[]
    sell=[]
    flag=-1
    for i in range(len(df)):
        if df['MACD'][i]>df['Signal_line'][i]:
            sell.append(np.nan)
            if flag !=1:
                buy.append(df['4. close'][i])
                flag=1
            else:
                buy.append(np.nan)
        elif df['MACD'][i]<df['Signal_line'][i]:
            buy.append(np.nan)
            if flag !=0:
                sell.append(df['4. close'][i])
                flag =0
            else:
                sell.append(np.nan)
        else:
            buy.append(np.nan)
            sell.append(np.nan)
    return(buy,sell)
                
                
```

```python
a=buy_sell(data)
a
data['buy_signal_price']=a[0]
data['sell_signal_price']=a[1]
```

```python
data
```

```python
plt.figure(figsize=(12.5,4.5))
plt.scatter(data.index,data['buy_signal_price'],label='sell',marker='^',color='red')
plt.scatter(data.index,data['sell_signal_price'],label='buy',marker='>',color='green')
plt.plot(data['4. close'],label='close_price',alpha=0.35,color='black')
#plt.title('close price buy and sell signal')
#plt.xlabel('date')
#plt.ylabel('close price')
plt.legend(loc='upper right')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(12.5,4.5))
plt.plot(data.index,MACD,label='MACD',color='red')
plt.plot(data.index,signal,label='Signal',color='blue')
#plt.plot(data['4. close'],label='close_price',alpha=0.35,color='black')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
```
